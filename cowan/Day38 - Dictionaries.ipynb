````python
# Solution to Lab 12: Tic Tac Toe

# Make a 2D list of all zeros with given dimensions.
def make2dList(rows, cols):
    g = []
    for row in range(0, rows): 
        g.append([0] * cols)
    return g

# Print the TTT board to the screen using X's and O's.
# Use the standard nested loops.
# Hint: Use print(...whatever..., end='') to
# not move to the next line at the end of a print statement.
# Hint 2: Use an empty print statement: print() to move the cursor
# to the next line.
def print_board(board):
    for r in range(len(board)):
        for c in range(len(board[r])):
            if board[r][c] == 1:
                print('X', end='')
            elif board[r][c] == -1:
                print("O", end='')
            else:
                print("-", end='')
        print()
    
# Add up and return the numbers in row r of the board.
# (Can do with a loop or manually.)
def add_row(board, r):
    total = 0
    for c in range(len(board[r])):
        total += board[r][c]
    return total        
    
# Add up and return the numbers in column c of the board.
# (Can do with a loop or manually.)
def add_column(board, c):
    total = 0

    for r in range(len(board)):
        total += board[r][c]

    return total
    
# Add up and return the numbers on the upper-left to
# lower-right of the board. (Can do with a loop or manually.)
def add_diagonal1(board):
    total = 0
    for row in range(0, len(board)):
        total += board[row][row]
    return total

# Add up and return the numbers on the upper-right to
# lower-left of the board. (Can do with a loop or manually.)
def add_diagonal2(board):
    total = 0
    for row in range(0, len(board)):
        col = len(board)-1 - row
        total += board[row][col]  

    return total
    
# Check if all the squares on the board are full (non-zero).
# Return True if they're all full, and False if they're not.
def board_full(board):
    for r in range(len(board)):
        for c in range(len(board[r])):
            #As soon as you find a zero, return false, since
            #board is not full (still spaces players can pick)
            if board[r][c] == 0:
                return False
              
    #If you get here, then must not have found any zeros, so return True
    return True

# Check to see if the sum of any row, column, or diagonal is 3.
# Return True for an X win, and False otherwise.
def x_wins(board):

    #Checks all rows for a win
    for r in range(len(board)):
        if add_row(board, r) == 3:
            return True
    #Checks all columns for a win
    for c in range(len(board[0])):
        if add_column(board, c) == 3:
            return True
    #Checks both diagonals
    if add_diagonal1(board) == 3:
        return True
    elif add_diagonal2(board) == 3:
        return True
      
    #If you reach here, then must not have won
    return False
    
    
# Check to see if the sum of any row or column is -3.
# Also check the diagonals for a win for O.
def o_wins(board):
    for r in range(len(board)):
        if add_row(board, r) == -3:
            return True
    for c in range(len(board[0])):
        if add_column(board, c) == -3:
            return True
    if add_diagonal1(board) == -3:
        return True
    elif add_diagonal2(board) == -3:
        return True

    return False

#This function returns True if game is over, False otherwise
#I added this function to make it easier to determine if game is over or not.
def game_over(board):
    if board_full(board) or x_wins(board) or o_wins(board):
        return True
    else:
        return False

def main():
    board = make2dList(3, 3)
    current_player = "X"

    while not game_over(board):
        
        print_board(board)
        print("Enter move for player", current_player)
        row = int(input("Row (0-2): "))
        col = int(input("Column (0-2): "))

        #Adds some input validation to ensure no one is trying to use a spot that has already been taken.
        while board[row][col] != 0:
            print("Spot already taken!")
            row = int(input("Row (0-2): "))
            col = int(input("Column (0-2): "))

        if current_player == 'X':
            board[row][col] = 1
            current_player = 'O'
        else:
            board[row][col] = -1
            current_player = 'X'

    print_board(board)
    if x_wins(board):
        print("X wins!")
    elif o_wins(board):
        print('O wins!')
    else:
        print("Tie game!")
    
main()
````

# Day 38: Dictionaries

__Dictionary:__ object that stores a collection of data
* Each element consists of a key and a value
* Often referred to as mapping of key to value
* Key must be an immutable object (cannot be changed!)
* To retrieve a specific value, use the key associated with it
* General Syntax: `dictionary = {key1:val1, key2:val2}`

* Keys must be immutable objects, but associated values can be any type of object
* One dictionary can include keys of several different immutable types
* Values stored in a single dictionary can be of different types


## Retrieving a Value from a Dictionary
* Elements in dictionary are unsorted
* General format for retrieving value from dictionary: `dictionary[key]`
* If key in the dictionary, associated value is returned, otherwise, KeyError exception is raised
* Test whether a key is in a dictionary using the in and not in operators (helps prevent KeyError exceptions)

````python
# Retrieving a Value from a Dictionary Example
def main():
  
    phonebook = {'Chris': '555-1111', 'Katie': '555-2222', 'JoAnne': '555-3333'}
  
    print(phonebook)
    print("Katie's phone number is: ", phonebook['Katie'])
    if 'James' in phonebook:
        print(phonebook['James'])
    else:
        print("not there")

main()
````

## Adding Elements to an Existing Dictionary

* Dictionaries are mutable objects
* To add a new key-value pair: `dictionary[key3] = value3`
* If key exists in the dictionary, the value associated with it will be changed

````python
# Adding Elements to an Existing Dictionary Example
def main():
  
    phonebook = {'Chris': '555-1111', 'Katie': '555-2222', 'JoAnne': '555-3333'}
  
    print("Before Add: ", phonebook)
  
    phonebook['Andy'] = '555-1234'
  
    print("After Add: ", phonebook)

main()
````

## Deleting Elements From an Existing Dictionary
* To delete a key-value pair: `del dictionary[key]`
* If key is not in the dictionary, KeyError exception is raised

````python
# Deleting Elements From an Existing Dictionary Example
def main():
  
    phonebook = {'Chris': '555-1111', 'Katie': '555-2222', 'JoAnne': '555-3333'}
  
    print("Before Delete: ", phonebook)
  
    del phonebook['JoAnne']
  
    print("After Delete: ", phonebook)

main()
````

## Dictionary Methods
Assume `D` is a dictionary; `k` and `v` are variables, either a number or a string.
* `len(D)`: returns the number of key-value pairs in D
* `k in D`: returns True if k is a key in dictionary D, False otherwise.
* `k not in D`: returns True if k is not a key in dictionary D, False otherwise
* `D[k] = v`: adds a new key-value pair to dictionary D, where k is the key and v is the value.
__Note:__ If k is already a key in D, then the previous value associated with k is overwritten with v.

* `D.get(k, default)`: returns the value associated with specified key (k) from the dictionary. If k is not in
the dictionary, it returns the default value (this is an alternative to using the [] operator, and it cannot raise KeyError exception)
* `D.keys()`: returns all the dictionaries keys as a sequence
* `D.values()`: returns all the dictionaries values as a sequence
* `del D[k]`: removes the key-value pair from dictionary D associated with key k.

````python
wordDictionary = {'the': 16, 'a': 7, 'whose': 1, 'its': 3}

keys = wordDictionary.keys()
print(keys)

for key in keys:
    print(key)
````
