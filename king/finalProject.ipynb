# Final Project

During finals week, you will present your final project to me. You **do not** need to prepare a formal presentation. You **do** need to understand what you have done and how you have done it.


### Total Time Expectations

This project replaces your final exam. I expect you to spend roughly 2-3 times the amount of time you normally spend on projects, including the proposal and presentation. Your project should require a **minimum** of 6 hours of work. I have a rough idea of what you can pull off in 6 hours, and we will negotiate a grade partially based on how much you have achieved in the expected time span.

### Project Proposal: **due Fri Nov 21 by email**

Send me an email with the subject **COMP 141 Final Project Proposal**. The proposal should include:
* a project title
* a brief project description
* a "minimal" list of features that you *know* you can implement
* an "expected" list of features that you *think* you can implement
* a "stretch" list of features that you *hope* you can implement

See the end for a few examples.

### Assessment

There **will not** be a final exam during our schedule exam time.

You **will** meet with me to present your project.

**Effort:**
* I will ask how long you worked. I expect you to be honest.
* I will compare your project with your "minimal" feature list. You should have achieved them all.
* I will compare your project with your "expected" feature list. I hope you've done some of them!
* I will ask you about your "stretch" list and how long you think it would take to implement. It's ok if you have done none of them.

**Knowledge:**
* I will ask you how you used **2D lists** in your project.
* I will ask you how you used **dictionaries** in your project.
* I will ask you how you used **functions** in your project.

### Requirements / Specifications

For a passing grade:
* You must spend at least 6 hours of honest, focused effort. Keep a record of time spent on different features/functions/research.
* You must use a reasonable number of functions for your skill level.
* You must implement a `main` function with an input validation loop.
* You must use at least one **dictionary** in your project.
* You must use at least one **2D list** in your project.

For a B:
* You must implement most of your "expected" feature list.

For an A:
* You must implemenet all of your "expected" feature list and at least one of your "stretch" features.

### Text-Based Adventure Game (w/ Graphics)

Design and implement a short, simple text-based adventure game. The player enters prompts into the text field (this would be the `main` game loop) related to what they want to do (`look around`, `fight`, `go to`, etc). There is one or several ways to win or lose. Dictionaries are used to store whether the player has unlocked certain keys or levels (for example). A 2D list is used as a kind of "minefield" minigame. The player can visit several locations and take several actions. Every "level" has a corresponding graphical scene.

### Sports (or other) Data Analysis

Design and implement a "sports statistics calculator", which takes in a single `txt` file (for a single player, single team, single season, etc) and computes various statistics (batting average, best time, worst game, etc). Every statistic should be a function. The `main` loop should let the user decide what file to open or what statistic to calculate.

### Image Editor

Build off of Project 8 to implement a more robust image editor. Changing pixel values is the way all image editors work. Possible functions include blacking out a circle or square of pixels, adding stripes to the image, and blurring the image by shuffling pixels around in a neighborhood.
