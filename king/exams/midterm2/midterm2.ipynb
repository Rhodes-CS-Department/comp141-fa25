### Question 3: Debugging

The following function `max_steps_collatz` is meant to do the following:
- out of all numbers up to some limit $n$: find the one that takes the most number of steps in its Collatz sequence before terminating.

It uses the helper function `collatz` which calculates the next number in a Collatz sequence.

The `collatz` function halves even numbers (so 18 becomes 9) and triples odd number and adds 1 (so 9 beomces 27 + 1 = 28).

It also uses the helper function `collatz_steps` which calculates the number of steps until the number becomes 1.

For example, 5 goes: 5 -> 16 -> 8 -> 4 -> 2 -> 1, which is 5 steps. Clearly, 16 takes 4 steps, 8 takes 3, 4 takes 2, and 2 takes 1 (1 takes 0).

Another example: 10 -> 5 -> ... -> 1. Since we know 5 takes 5 steps, 10 then takes 6 steps.

`collatz` **works fines**. Do not modify it.

`collatz_steps` **is broken!**. Can you fix it?

`max_steps_collatz` **is broken!** Can you fix it?

```python
# Part 1: Make sure the program runs without error.
# Part 2: Make sure the program does what it's supposed to.

def collatz(n): # this function is correct: do not modify it
    if n % 2 == 0:
        return n // 2
    else:
        return 3 * n  + 1


def collatz_steps(n): # this is broken! it should count the number of steps it takes for a number n to turn into 1.
    next = n
    steps = 0
    while next != 1
        # hint: if you're very confused, try printing next! just don't leave the print statement in your final submission!
        # you should see a valid collatz sequence, like 9, 28, 14, 7, 22, 11, ...
        next = collatz(next)
        steps += 1
    print(steps)


def max_steps_collatz(n): # this is broken! it should find the maximum number of steps for all numbers from 1 to n.
    max_steps = 0
    max_steps_number = 0
    for number in range(n)
        steps = collatz_steps(number)
        if steps < max_steps:
            max_steps = steps
            max_steps_number = number
    return max_steps_number, max_steps # this line is correct!


# test collatz_steps
print(collatz_steps(9)) # should be 19
print(collatz_steps(20)) # should be 7
print(collatz_steps(100)) # should be 25

# test max_steps_collatz
print(max_steps_collatz(9)) # should print (9, 19), since 9 takes 19 steps before becoming 1: 9 -> 28 -> 14 -> 7 -> 22 -> 11 ...
print(max_steps_collatz(20)) # should print (18, 20), since 18 takes 20 steps before becoming 1: 18 -> 9 -> 28 (see above)
print(max_steps_collatz(100)) # should print (97, 118)
```
