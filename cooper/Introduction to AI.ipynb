## A Neural Network is a Data Pipeline

Each connection between nodes has a <b>weight</b> associated with it. Each node also has an internal paramter called its <b>bias</b>. In a neural network, the weights and biases are collectively referred to as its <b>parameters</b>.

Training a Neural Network:
- Start with random weights
- Make a prediction
- Compare with the correct answer
- Adjust weights
- Repeat many times


## Training

How wrong were we? Use Mean Squared Error:

Loss = (1/n) * Σ (prediction - truth)^2

A common way to train a neural network is to use <b>gradient descent</b>, also referred to as <b>stochastic hill climbing</b>. The algorithm when implemented for a neural network is called <b>backpropagation</b>. Basically, it's a way to measure how much each weight contributed to the error. The weights are then adjusted slightly to try to fix the error. Try again. A whole bunch of times.

Error → Output Layer → Hidden Layer → Input Layer
         (push backward corrections)



## Recurrent Network

The output of each hidden layer node is fed back as input in the next iteration. This allows the network to "remember" the previous state.

## When Neural Networks Fail

Neural networks struggle with:

- Small or biased data
- Poor labels
- Noisy inputs (bad data)
- “Edge cases”

Neural networks have inherent limitations:

- Need huge datasets
- Difficult to explain
- Slow to train
- Can be fooled by small changes
- Energy-intensive
